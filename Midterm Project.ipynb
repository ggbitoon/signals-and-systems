## 1. From the Fourier series for the periodic function f(t), create a python program that will ask the user to input the value of the fundamental period T and generate the plot for the periodic function f(t). Note that the plot is only an approximation and the level of fitness depends on the number of iteration and resolution of the data. (5 pts) {-}

```python
def fourier_transform_1(T):
    t_values = np.linspace(-T, T, 1000)
    n = np.arange(1, 2000)
    y = []
    
    for i in t_values:
        fn = 0
        
        for j in n:
            fn += (2/(j*np.pi))*(1-np.cos(j*np.pi))*(np.sin((2*j*np.pi/T)*i))
        
        y.append(fn)
        
    return t_values, y
```

```python
def fourier_plot_1(T):
    
    xlist = fourier_transform_1(T)[0]
    ylist = fourier_transform_1(T)[1]
    
    plt.plot(xlist,ylist)
    plt.title("Fourier Series")
    plt.grid(True)
    plt.show()
```

```python
print('Enter the Fundamental Period T')
T = float(input())
fourier_plot_1(T)
```

## 2. From the Fourier series for the periodic function f(t), create a python program that will generate the plot of the periodic function f(t). Note that the plot is only an approximation and the level of fitness depends on the number of iteration and resolution of the data. (5 pts.). {-}

```python
def fourier_transform_2():
    t_values = np.linspace(-2*np.pi, 2*np.pi, 100)
    n = np.arange(1, 2000)
    y = []
    
    for i in t_values:
        
        fn1 = 0
        fn2 = 1/4
        
        for j in n:
            fn1 += (1/((j**2)*(np.pi**2)))*(j*np.pi*np.sin(j*np.pi)*np.cos(j*np.pi) + np.cos(j*np.pi)*np.cos(j*i) - np.cos(j*i))
            fn2 += (np.sin(j*np.pi)*np.sin(j*i)/((j**2)*(np.pi**2))) - (np.cos(j*np.pi)*np.sin(j*i)/(j*np.pi))
        
        y.append(fn1 + fn2)
        
    return t_values, y
```

```python
def fourier_plot_2():
    
    xlist = fourier_transform_2()[0]
    ylist = fourier_transform_2()[1]
    
    plt.plot(xlist,ylist)
    plt.title("Fourier Series")
    plt.grid(True)
    plt.show()
```

## 4. Create a code that will compute the discrete Fourier transform of $𝑥[𝑛] = 0.5^𝑛𝑢[𝑛]$, where u[n] is the unit step sequence. Plot the real part and imaginary part of the discrete Fourier transform  {-}

```python
def number_4(a):
    
    x = np.linspace(-np.pi, np.pi, 1000)
    
    y = []
    y_re = []
    y_im = []
    
    for i in x:
        y.append((1/(1-a*(np.cos(i)-np.sin(i)*1j))))
    
    for i in y:
        y_re.append(i.real)
        y_im.append(i.imag)
    
    return x, y_re, y_im
```

```python
def number_4_plot(a):
    
    z = number_4(a)
    
    plt.plot(z[0],z[1])
    plt.plot(z[0],z[2])
    plt.grid()
    plt.show()
```
