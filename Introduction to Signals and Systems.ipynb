#### 1. create a function for Kronecker-delta in Python, Matlab, or Sci-lab. plot: {-}
#### $y[n] = \delta[n-5]$ {-}

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def kron_delta_func(m):
    
    n = np.arange(-5,10)
    y = []
    x = len(n)
    
    for i in range(x):
        
        if n[i] == m:
            y.append(1)
            
        else:
            y.append(0)
            
    plt.stem(n, y, use_line_collection = True)    
    plt.title('Kronecker Delta Function')
    plt.xlim(n[0]-1, n[-1]+1)
    plt.xticks(np.arange(n.min(), n.max(), 1))
    plt.grid()
    plt.show()
    
    return n, y
```

```python
number_1_a = kron_delta_func(5)
```

#### 2. create a function for unit step sequence in Python, Matlab, or Sci-lab. plot: {-}
#### $y[n] = H[n-5]$ {-}

```python
def unit_step_func(m, a):
    
    n = np.arange(-3,15)
    y = []
    x = len(n)
    
    for i in range(x):
        
        if n[i] >= m:
            y.append(a**(i)*1)
            
        else:
            y.append(0)

    plt.stem(n, y, use_line_collection = True)    
    plt.title('Unit Step Function')
    plt.xlim(n[0]-1, n[-1]+1)
    plt.xticks(np.arange(n.min(), n.max(), 1))
    plt.grid()
    plt.show()
    
    return n, y
        
```

```python
number_1_b = unit_step_func(7, 1)
```

#### c.  plot the output $y[n]$ by using and making a Python, Matlab, or Sci-lab program for the convolution sum {-}
#### $y[n]=x[n]*I[n]=\sum^{\infty}_{k=-\infty}{x[k]I[n-k]}$ {-}


```python
def number_2_c(a):
    
    n = np.arange(-5,10)
    y = []
    output = 0
    
    for i in n:
        
        z = 0
        if i >= a:
            z = 1
        else:
            z = 0
            
        output += (0.5**i)*z**2
        y.append(output)
        

    plt.stem(n, y, use_line_collection = True)    
    plt.title('Plot of $y[n]$')
    plt.xlim(n[0]-1, n[-1]+1)
    plt.xticks(np.arange(n.min(), n.max(), 1))
    plt.grid()
    plt.show()
```

#### after manipulating the equation, we obtain: $y[n]=\sum^n_{m=0}{\alpha^m}=\frac{1-\alpha^{n+1}}{1-\alpha}$ {-}

```python
def number_2_d(a):
    
    n = np.arange(0,10)
    y = []
    output = 0
    
    for i in n:
        output = (1-(0.5**(i+1)))/(1-0.5)
        y.append(output)
        
    plt.stem(n, y, use_line_collection = True)    
    plt.title('Plot of $y[n]$')
    plt.xlim(n[0]-1, n[-1]+1)
    plt.xticks(np.arange(n.min(), n.max(), 1))
    plt.grid()
    plt.show()
```

#### $x[n] = \delta[n] + \delta[n-1] + \delta[n-2] + \delta[n-3]$ {-}
#### $I[n] = \delta[n] + \delta[n-1] + \delta[n-2]$ {-}
#### $x[n] * I[n] = y[n]$ {-}
#### $x[n]I[n] = x[n](\delta[n]+\delta[n-1]+\delta[n-2])$ {-}
#### $x[n]I[n] = x[n]\delta[n] + x[n]\delta[n-1] + x[n]\delta[n-2]$ {-}
#### $\bf{x[n]*I[n] = y[n] = \delta[n]+2\delta[n-1]+3\delta[n-2]+3\delta[n-3]+2\delta[n-4]+\delta[n-5]}$ {-}

#### b. create a program for the convolution sum then plot the output $y[n]$ {-}

```python
def kron(n, a):
    if n == a:
        return 1
    else:
        return 0

def number_3_b():
    
    n = np.arange(-5,10)
    y = []
    
    for i in n:
        
        output = kron(i, 0) + 2*kron(i, 1) + 3*kron(i, 2) + 3*kron(i, 3) + 2*kron(i, 4) + kron(i, 5)
        y.append(output)

    plt.stem(n, y, use_line_collection = True)    
    plt.title('Plot of $y[n]$')
    plt.xlim(n[0]-1, n[-1]+1)
    plt.xticks(np.arange(n.min(), n.max(), 1))
    plt.grid()
    plt.show()
```
