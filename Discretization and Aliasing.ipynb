## 1. Discretize the following sine wave signals with 100 Hz sampling rate: {-}
### $x_c(t)=\sin{2\pi f t}$ where $f$(in Hz)$=10,90,20,80,36,64,49,51$ {-}

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def number_1(f, fs):
    n = np.linspace(-25, 25, fs)
    y = []
    Ts = 1/fs
    
    for i in range(len(n)):
        
        fn = np.sin((2*(np.pi)*f*n[i]*Ts))
        y.append(fn)
        n[i] = n[i] * Ts
    
    return n, y

def number_1_plot(f1, f2, f3, f4, f5, f6, f7, f8, fs):
    
    z = [f1, f2, f3, f4, f5, f6, f7, f8]
    
    for i in z:
        
        xlist = number_1(i, fs)[0]
        ylist = number_1(i, fs)[1]
        
        
        plt.plot(xlist, ylist)
        plt.title('f = {}'.format(i))
        plt.scatter(xlist, ylist)
        plt.xlabel('t=nT')
        plt.ylabel('x')
        plt.show()
    
```

```python
number_1_plot(10, 90, 20, 80, 36, 64, 49, 51, 100)
```

## Question #1: What do you observe? {-}
### There are certain signals that seem discretized properly, while others show erratic behavior. There seems to be a frequency limit where sampling is accurate. {-}

## Question #2: What is the maximum signal frequency representable by 100Hz sampling rate? {-}
### The maximum signal frequency is 1/2 the sampling rate - in this case, 50 Hz. {-}

## Question #3: What happens if you try to sample a signal above this maximum? {-}
### Aliasing occurs when a signal is sampled above the maximum signal frequency. The signals would then be indistinguishable. {-}
